```python
#Ques 3 : What errors are defined in the ArithmeticError class? Explain any two with an example.

"""Answer : ArithmeticError Exception is the base class for all errors that occur for numeric calculations. It is the base class for those built-in exceptions like: OverflowError, ZeroDivisionError, FloatingPointError"""

try:
    7/0
except ArithmeticError as e:
    print ("This is an example of catching ArithmeticError " , e)
```

```python
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print("Overflow error happened",e)
```

```python
#Ques 4 : Why LookupError class is used? Explain with an example KeyError and IndexError.

"""LookupError Exception is the Base class for errors raised when something can’t be found. The base class for the exceptions that are raised when a key or index used on a mapping or sequence is invalid: IndexError, KeyError.

An IndexError is raised when a sequence reference is out of range."""

try:
    foo = ['a', 's', 'd', 'f', 'g']
    print(foo[5])
except IndexError as e:
    print ('error Index ',e)
```

```python
"""The Python KeyError is an exception that occurs when an attempt is made to access an item in a dictionary that does not exist. """

try:
    ages = {1: 30, 2: 28, 3: 33}
    print(ages[4])
except KeyError as e:
    print("KeyError ",e)
```

```python
#Ques 5 : Explain ImportError. What is ModuleNotFoundError?

"""Answer : ImportError occurs when the Python program tries to import module which does not exist in the private table"""
try:
    import hemant

    print("hello")
except ModuleNotFoundError as e:
    print("Imported module not found" , e)
```
